# Practical 2: Mini Calculator

In the example below we use the **float()** function to convert the first and second number we receive from the user to float data type, so that we work with floats.

```python
# get first number from the user. The user would type this number using the keyboard
num1 = float(input('Enter first number: '))
# get second number from the user.
num2 = float(input('Enter second number: '))

print('Sum:', num1 + num2)
print('Difference:', num1 - num2)
print('Product:', num1 * num2)
print('Quotient:', num1 / num2)
```

# Practical 3: Calculator with Operation Selection

Notice that we use **if** and **elif** statement in python, so that our program can make decisions based on the value of the **op** variable we receive from the user.

```python
num1 = float(input('Enter first number: '))
num2 = float(input('Enter second number: '))
op = input('Enter operation (+, -, *, /): ')

if op == '+':
   print(num1 + num2)
elif op == '-':
    print(num1 - num2)
elif op == '*':
    print(num1 * num2)
elif op == '/':
    print(num1 / num2)
else:
    print('Invalid operation')
```

# Practical 4: Reading CSV File

As a Data Scientist or Data Analyst you would usually work with a dataset.

A **dataset** is a structured collection of related data, typically organized in rows and columns (tabular format).

Popular formats for dataset files are **CSV** and **JSON**

**CSV** means **Comma-Separated Values**
**JSON** means **JavaScript Object Notation**

The code below shows how you would read a CSV file in python using the open command.

```python
# open the csv file for the python script to read
file = open('DTS101_week2_student_scores.csv', 'r')

count = 0
# loop through and print out each line in the opened file for us to see
for line in file:
    count += 1
    print("line ", count, ": --> ", line)

    if count == 10:
        break

# close the file after we are done with our file operations
file.close()
```

# Practical 5: Extract Math Scores and Compute Average

```python
# open the csv file for the python script to read
file = open('DTS101_week2_student_scores.csv', 'r')

# initial variables
total = 0
count = 0

# The next() function in Python retrieves the next item from an iterator
# for now, think of this line as us skipping the heading of our csv file
# (that contains student_id,math,english,physics,chemistry)
next(file)

# loop through the opened file
for line in file:
    # split the lines into different parts using the ',' as a delimiter
    parts = line.strip().split(',')
    # remember the math is at position 0
    math_score = int(parts[1])
    # calulate the total for math score
    total += math_score
    # count the line
    count += 1

# calculate and print the average for the math score
average = total / count
print('Average math score:', average)

# close the file
file.close()
```
