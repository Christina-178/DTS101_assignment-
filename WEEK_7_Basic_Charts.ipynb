# PART 1: Importing Matplotlib

```python
import matplotlib.pyplot as plt
```

# PART 2: Create Sample Dataset

```python

students = ["Alice", "Bob", "Charlie", "David", "Eva"]
math_scores = [65, 80, 45, 90, 75]
english_scores = [70, 85, 50, 88, 80]
```

# PART 3: Line Chart (Math Scores)

```python

plt.figure()
plt.plot(students, math_scores)
plt.title("Math Scores - Line Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

# PART 4: Bar Chart (Math Scores)

```python

plt.figure()
plt.bar(students, math_scores)
plt.title("Math Scores - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

# PART 5: Pie Chart (Math Score Distribution)

```python

plt.figure()
plt.pie(math_scores, labels=students, autopct='%1.1f%%')
plt.title("Math Score Distribution")
plt.show()
```

# PART 6: Compare Math and English (Separate Charts)

```python

plt.figure()
plt.plot(students, english_scores)
plt.title("English Scores - Line Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```
