```python

df["Average"] = df[["Math", "English", "Science"]].mean(axis=1)
df
```

```python

df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 50 else "Fail")
df
```

#1. Student file

```python
import pandas as pd
```

```python


data = {
"Name": ["Mikasa", "Eren", "Levi", "Armin", "Kenny", "Reiner", "Yemir", "Sasha", "Hange", "Zeke", "Conny", "Jean"],
"Math": [80, 96, 89, 90, 56, 70, 65, 89, 90, 45, 35, 78],
"English": [90, 80, 77, 86, 97, 67, 77, 56, 90, 78, 54, 64],
"Computer": [88, 90, 99, 76, 86, 90, 77, 98, 88, 99, 79, 91]
}
df = pd.DataFrame(data)

df.to_csv("Students.csv", index=False)
```

#2. Load the csv

```python

df = pd.read_csv("Students.csv")
df
```

#3. Compute :
>Subject average
>Overall average

>Subject_averages

```python
subject_averages = df[["Math", "English", "Computer"]].mean()
print("Average")
print(subject_averages)
```

>Overall_average

```python
df["Average"] = df[["Math", "English", "Computer"]].mean(axis=1)
df
```

```python
overall_average = df["Average"].mean()
print("Overall Average", overall_average)
```

#4. Identify
>Top performing student

>Lowest performing student

>Top performing student

```python
highest_avg = df["Average"].max()
top_student = df[df["Average"] == highest_avg]
print("Top performing student")
print(top_student)
```

>Lowest performing student

```python
lowest_avg = df["Average"].min()
lowest_student = df[df["Average"] == lowest_avg]
print("Lowest Performing student")
print(lowest_student)
```
