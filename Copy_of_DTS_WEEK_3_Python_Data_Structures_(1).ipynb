# PART 1: LISTS

```python
scores = [65, 70, 80, 55, 90]
print(scores)
```

```python
print(scores[0])
print(scores[2])
```

```python
for score in scores:
    print(score)
```

```python
total = 0
for score in scores:
    total = total + score

average = total / len(scores)
print("Average score:", average)
```

```python
highest = scores[0]
for score in scores:
    if score > highest:
        highest = score

print("Highest score:", highest)
```

# PART 4: DICTIONARIES

```python
student = {
    "name": "David",
    "math": 75,
    "english": 80,
    "science": 70
}

print(student)
print(student["math"])
```

```python
for key in student:
    print(key, student[key])
```

```python
total = 0
for subject in ["math", "english", "science"]:
    total += student[subject]

average = total / 3
print("Average:", average)
```

# PART 5: Multiple Student Records

```python
students = {
    "S001": 65,
    "S002": 80,
    "S003": 45,
    "S004": 90
}

print(students)
```

```python
highest_id = None
highest_score = 0

for student_id in students:
    score = students[student_id]

    if score > highest_score:
        highest_score = score
        highest_id = student_id

print("Highest scoring student:", highest_id)
print("Score:", highest_score)
```
