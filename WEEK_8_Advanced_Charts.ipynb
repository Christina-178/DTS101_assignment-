# PART 1: Import Libraries

```python

import matplotlib.pyplot as plt
import numpy as np
```

# PART 2: Create Sample Dataset

```python

students = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen", "Ian", "Jane"]
math_scores = [65, 80, 45, 90, 75, 55, 88, 92, 60, 73]
english_scores = [70, 85, 50, 88, 80, 58, 90, 95, 62, 77]
science_scores = [72, 78, 48, 92, 79, 60, 85, 98, 64, 74]
```


# PART 3: Histogram

A histogram shows the distribution of numerical data.


```python

plt.figure()
plt.hist(math_scores, bins=5)
plt.title("Distribution of Math Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```


# PART 4: Scatter Plot

A scatter plot shows the relationship between two variables.


```python

plt.figure()
plt.scatter(math_scores, english_scores)
plt.title("Math vs English Scores")
plt.xlabel("Math Scores")
plt.ylabel("English Scores")
plt.show()
```


# PART 5: Multiple Line Chart

Compare trends of multiple subjects.


```python

plt.figure()
plt.plot(students, math_scores)
plt.plot(students, english_scores)
plt.title("Math and English Scores Comparison")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```


# PART 6: Subplots

Display multiple charts in one figure.


```python

plt.figure()

plt.subplot(1, 2, 1)
plt.hist(science_scores, bins=5)
plt.title("Science Histogram")

plt.subplot(1, 2, 2)
plt.scatter(math_scores, science_scores)
plt.title("Math vs Science")

plt.tight_layout()
plt.show()
```

#1. Student Data set

```python
students = ["Kane", "Barbara", "Charles", "Dean", "Everest", "Duglas", "Abraham", "Daina", "Max", "Dustin", "Lucas", "Will", "Jane", "Mike", "Steve", "Vecna"]
english_scores = [87, 89, 99, 79, 97, 99, 79, 97, 90, 85, 95, 80, 92, 77, 98, 87]
computer_scores = [90, 85, 95, 80, 92, 89, 99, 79, 97, 99, 79, 97, 90, 85, 95, 77]
```

#2, 3.
°Histogram

°Scatter plot

°Multiple line chat

°Sub plots

```python
#Histogram


plt.figure()
plt.hist(english_scores, bins=5, alpha=0.7, label='English', color='orange', edgecolor='black')
plt.hist(computer_scores, bins=5, alpha=0.7, label='Computer', color='red', edgecolor='black')

plt.title("Histogram of English and Computer Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```

```python
#scatter plot


plt.figure(figsize=(8,5))
plt.scatter(students, english_scores, color='indigo', label='English', marker='o')
plt.scatter(students, computer_scores, color='maroon', label='Computer', marker='s')

plt.title("Scatter Plot of English and Computer Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.xticks(rotation=45)  # Rotate student names for better visibility
plt.legend()  # Add legend to distinguish subjects
plt.grid(True)  # Optional: add grid for clarity
plt.show()
```

```python
#. Multiple line chart


plt.figure()
plt.plot(students, english_scores, marker='o', label="English")
plt.plot(students, computer_scores, marker='x', label="Computer")

plt.title("English vs Computer Scores")
plt.xlabel("Students")
plt.ylabel("Scores")

plt.legend()

plt.show()
```

```python
#sub plot


plt.figure(figsize=(10,5))
plt.plot(students, english_scores, color='green', marker='o', linestyle='-', label='English')
plt.plot(students, computer_scores, color='brown', marker='s', linestyle='--', label='Computer')

plt.title("Combined Line Chart of English and Computer Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

In the Histogram, it shows that most students scored between 70 - 90. In the scatter plot it shows Charles is the highest scorer in English. In Computer, both Dustin and Abraham are the highest scorers in Computer. In general the highest scoring student in English is Charles. Whereass in computer there are two students with the highest scores, which are Abraham and Dustin. Generally, every student scored above high marks.
