
# PART 3: Function to Compute Average

```python

def compute_average(scores):
    total = 0

    for score in scores:
        total = total + score

    average = total / len(scores)

    return average


scores = [65, 70, 80, 90, 75]

avg = compute_average(scores)

print("Average score:", avg)
```


# PART 4: Function to Determine Pass or Fail


```python

def check_pass_fail(score):

    if score >= 50:
        return "Pass"
    else:
        return "Fail"


print(check_pass_fail(75))
print(check_pass_fail(40))
```


# PART 5: Apply Function to Multiple Students


```python

scores = [45, 67, 80, 30, 90]

for score in scores:
    result = check_pass_fail(score)
    print(score, result)
```
